##4h (2250 trabalhos)

```python
lista_colecoes = joblib.load(r'/content/drive/MyDrive/Programa - Repositório Institucional UFSC/Extração de Dados/via API DSpace Protocolo OAI-PMH/Etapa 1/Lista de colecoes/lista_de_colecoes.joblib')


dic_listagem_execs = {'Dia 1':{'Programa 1':[0,[]],'Programa 2':[0,[]],'Programa 3':[0,[]],'Programa 4':[0,[]],'Programa 5':[0,[]],'Programa 6':[0,[]]},
                      'Dia 2':{'Programa 1':[0,[]],'Programa 2':[0,[]],'Programa 3':[0,[]],'Programa 4':[0,[]],'Programa 5':[0,[]],'Programa 6':[0,[]]},
                      'Dia 3':{'Programa 1':[0,[]],'Programa 2':[0,[]],'Programa 3':[0,[]],'Programa 4':[0,[]],'Programa 5':[0,[]],'Programa 6':[0,[]]},
                      'Dia 4':{'Programa 1':[0,[]],'Programa 2':[0,[]],'Programa 3':[0,[]],'Programa 4':[0,[]],'Programa 5':[0,[]],'Programa 6':[0,[]]},
                      'Dia 5':{'Programa 1':[0,[]],'Programa 2':[0,[]],'Programa 3':[0,[]],'Programa 4':[0,[]],'Programa 5':[0,[]],'Programa 6':[0,[]]},
                      'Dia 6':{'Programa 1':[0,[]],'Programa 2':[0,[]],'Programa 3':[0,[]],'Programa 4':[0,[]],'Programa 5':[0,[]],'Programa 6':[0,[]]}}

indices_usados = []
for dia in dic_listagem_execs.keys():
  for programa in dic_listagem_execs[dia]:
    for i in range(len(lista_colecoes)):
      if i not in indices_usados:
        if dic_listagem_execs[dia][programa][0] < 1900:
          if dic_listagem_execs[dia][programa][0] + lista_colecoes[i][2] < 2200:
            dic_listagem_execs[dia][programa][0] += lista_colecoes[i][2]
            dic_listagem_execs[dia][programa][1].append(lista_colecoes[i][0])
            indices_usados.append(i)
          elif lista_colecoes[i][2] > 2200:
            if dic_listagem_execs[dia][programa][0] == 0:
              dic_listagem_execs[dia][programa][0] += lista_colecoes[i][2]
              dic_listagem_execs[dia][programa][1].append(lista_colecoes[i][0])
              indices_usados.append(i)

contagem = 0
for dia in dic_listagem_execs.keys():
  print(dia)
  for programa in dic_listagem_execs[dia].keys():
    print(programa,dic_listagem_execs[dia][programa])
    contagem += dic_listagem_execs[dia][programa][0]
  print('\n')

print(contagem)

for i in range(len(lista_colecoes)):
  if i not in indices_usados:
    print(lista_colecoes[i][0])

joblib.dump(dic_listagem_execs,r'/content/drive/MyDrive/Programa - Repositório Institucional UFSC/Extração de Dados/via API DSpace Protocolo OAI-PMH/Etapa_1_2/Dicionário de Execuções/dic_listagem_execs.joblib')
```

Texto para o Docs:

```python
texto = 'Listagem das coleções no site do repositório:\n\n'

for colecao in lista_colecoes:
  texto += colecao[0]+' ['+str(colecao[2])+']\n'

texto += '\n\nDivisão de coleções por dia e programa:\n\n'
for dia in dic_listagem_execs.keys():
  texto += dia
  for programa in dic_listagem_execs[dia].keys():
    texto += '\n'+programa+': '+'['+str(dic_listagem_execs[dia][programa][0])+']'
    for item in dic_listagem_execs[dia][programa][1]:
      texto+='\n'+item
    texto+='\n'
  texto+='\n\n'

with open('Texto.txt','w',encoding='utf-8') as f:
  f.write(texto)
  f.close()
```
