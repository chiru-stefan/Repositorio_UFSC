```python
# Sem exclusão
# def TextosPProcessDatas(caminho_colecao : str, ano1 : int, ano2 : int):
#   lista_de_anos_desejados = []
#   ano = ano1
#   while ano <= ano2:
#     lista_de_anos_desejados.append(str(ano))
#     ano += 1
#   textos_pre_processados_intervalo_datas = []
#   # pasta_anos = [elemento for elemento in os.listdir(caminho_colecao) if (elemento.isdigit()) and (elemento in lista_de_anos_desejados)]
#   # for ano_desejado in pasta_anos:
#   for ano_desejado in lista_de_anos_desejados:
#     print('Passando pelo ano:',ano_desejado)
#     if os.path.exists(os.path.join(caminho_colecao,str(ano_desejado),'textos_pre_processados_ano.joblib')):
#       print('\nColetando e juntando textos pré-processados desse ano...')
#       textos_pre_processados_intervalo_datas += joblib.load(os.path.join(caminho_colecao,str(ano_desejado),'textos_pre_processados_ano.joblib'))
#   return textos_pre_processados_intervalo_datas

# Com exclusão do segundo ano fornecido
# def TextosPProcessDatas(caminho_colecao : str, ano1 : int, ano2 : int):
#   lista_de_anos_desejados = []
#   ano = ano1
#   while ano < ano2:
#     lista_de_anos_desejados.append(str(ano))
#     ano += 1
#   textos_pre_processados_intervalo_datas = []
#   # pasta_anos = [elemento for elemento in os.listdir(caminho_colecao) if (elemento.isdigit()) and (elemento in lista_de_anos_desejados)]
#   # for ano_desejado in pasta_anos:
#   for ano_desejado in lista_de_anos_desejados:
#     print('Passando pelo ano:',ano_desejado)
#     if os.path.exists(os.path.join(caminho_colecao,str(ano_desejado),'textos_pre_processados_ano.joblib')):
#       print('\nColetando e juntando textos pré-processados desse ano...')
#       textos_pre_processados_intervalo_datas += joblib.load(os.path.join(caminho_colecao,str(ano_desejado),'textos_pre_processados_ano.joblib'))
#   return textos_pre_processados_intervalo_datas

def TextosPreProcessadosDatasComRE(caminho_colecao : str, ano1 : int, ano2 : int):
  lista_de_anos_desejados = []
  ano = ano1
  while ano <= ano2:
    lista_de_anos_desejados.append(str(ano))
    ano += 1
  textos_pre_processados_intervalo_datas = []

  for ano_desejado in lista_de_anos_desejados:
    print('Passando pelo ano:',ano_desejado)
    if os.path.exists(os.path.join(caminho_colecao,str(ano_desejado),f'texto_pre_processado_com_re_{ano_desejado}.joblib')):
      print('\nColetando e juntando textos pré-processados desse ano...')
      textos_pre_processados_intervalo_datas += joblib.load(os.path.join(caminho_colecao,ano_desejado,f'texto_pre_processado_com_re_{ano_desejado}.joblib'))
  return textos_pre_processados_intervalo_datas

def TextosPreProcessadosDatasSemRE(caminho_colecao : str, ano1 : int, ano2 : int):
  lista_de_anos_desejados = []
  ano = ano1
  while ano <= ano2:
    lista_de_anos_desejados.append(str(ano))
    ano += 1
  textos_pre_processados_intervalo_datas = []

  for ano_desejado in lista_de_anos_desejados:
    print('Passando pelo ano:',ano_desejado)
    if os.path.exists(os.path.join(caminho_colecao,str(ano_desejado),f'texto_pre_processado_sem_re_{ano_desejado}.joblib')):
      print('\nColetando e juntando textos pré-processados desse ano...')
      textos_pre_processados_intervalo_datas += joblib.load(os.path.join(caminho_colecao,ano_desejado,f'texto_pre_processado_sem_re_{ano_desejado}.joblib'))
  return textos_pre_processados_intervalo_datas


def treinaModelo(corpus : list):
  model = Word2Vec(sentences=corpus,vector_size=100,alpha=0.025,negative=5,window=5,sample=1e-3,epochs=5,min_count=35,workers=2,sg=0)
  return model

def criaDiretorio(caminho : str, pasta: str):
  caminho_atualizado = os.path.join(caminho,pasta)
  if not os.path.exists(caminho_atualizado):
    try:
      os.makedirs(caminho_atualizado)
      time.sleep(1)
      if not os.path.exists(caminho_atualizado):
        time.sleep(2)
        os.makedirs(caminho_atualizado)
        time.sleep(3)
        if not os.path.exists(caminho_atualizado):
          return False
        else:
          return True
      else:
        print(f'Diretório "{caminho_atualizado}" criado com sucesso.')
        return True
    except:
      print(f'Erro ao criar diretorio:"{caminho_atualizado}".\nDescrição do erro --> {str(sys.exc_info()[0].__name__)}: {str(sys.exc_info()[1])}')
      return False
  else:
    return True
```
