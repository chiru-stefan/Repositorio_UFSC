##Palavras mais próximas

```python
# Escolha a palavra, dimensão e o número de resultados (topn) e execute esta célula
palavra = 'raça'
dimensao = 400
topn = 20

if dimensao == 400:
  print('WOKE SKIP-GRAM')
  try:
    for resultado in word2vec_woke_hst_sg_400.most_similar(palavra,topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')
  print('\n')

  print('WOKE CBOW')
  try:
    for resultado in word2vec_woke_hst_cbow_400.most_similar(palavra,topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')
  print('\n')

  print('NONCE2VEC SKIP-GRAM')
  try:
    for resultado in nonce2vec_woke_hst_sg_400.most_similar(palavra,topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')
  print('\n')

  print('NONCE2VEC CBOW')
  try:
    for resultado in nonce2vec_woke_hst_cbow_400.most_similar(palavra,topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')


elif dimensao == 700:
  print('WOKE SKIP-GRAM')
  try:
    for resultado in word2vec_woke_hst_sg_700.most_similar(palavra,topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')
  print('\n')

  print('WOKE CBOW')
  try:
    for resultado in word2vec_woke_hst_cbow_700.most_similar(palavra,topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')

elif dimensao == 1000:
  print('WOKE SKIP-GRAM')
  try:
    for resultado in word2vec_woke_hst_sg_1000.most_similar(palavra,topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')
  print('\n')

  print('WOKE CBOW')
  try:
    for resultado in word2vec_woke_hst_cbow_1000.most_similar(palavra,topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')
```

##Comparação

```python
# Escolha os valores para serem substituídos por homem, rei e mulher, dimensão e número de resultados (topn) e execute a célula

# HOMEM --> REI
# MULHER --> ?

homem = 'história'
rei = 'passado'
mulher = 'tecnologia'
dimensao = 400
topn = 50

if dimensao == 400:
  print('WOKE SKIP-GRAM')
  try:
    for resultado in word2vec_woke_hst_sg_400.most_similar_cosmul(positive=[rei,mulher], negative=[homem],topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')
  print('\n')

  print('WOKE CBOW')
  try:
    for resultado in word2vec_woke_hst_cbow_400.most_similar_cosmul(positive=[rei,mulher], negative=[homem],topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')
  print('\n')

  print('NONCE2VEC SKIP-GRAM')
  try:
    for resultado in nonce2vec_woke_hst_sg_400.most_similar_cosmul(positive=[rei,mulher], negative=[homem],topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')
  print('\n')

  print('NONCE2VEC CBOW')
  try:
    for resultado in nonce2vec_woke_hst_cbow_400.most_similar_cosmul(positive=[rei,mulher], negative=[homem],topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')

elif dimensao == 700:
  print('SKIP-GRAM')
  try:
    for resultado in word2vec_woke_hst_sg_700.most_similar_cosmul(positive=[rei,mulher], negative=[homem],topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')
  print('\n')

  print('CBOW')
  try:
    for resultado in word2vec_woke_hst_cbow_700.most_similar_cosmul(positive=[rei,mulher], negative=[homem],topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')

elif dimensao == 1000:
  print('SKIP-GRAM')
  try:
    for resultado in word2vec_woke_hst_sg_1000.most_similar_cosmul(positive=[rei,mulher], negative=[homem],topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')
  print('\n')

  print('CBOW')
  try:
    for resultado in word2vec_woke_hst_cbow_1000.most_similar_cosmul(positive=[rei,mulher], negative=[homem],topn=topn):
      print(resultado)
  except:
    print('Erro ao retornar resultados')
```
